3.Describe three different data types.

Ans --->

1)Dictionaries(dict) ---> Dictionaries are unorderd mappings for storing objects. Dictionaries uses a key-value pair.
This key-value pair allows users to quickly grab objects without needing to know an index location. Dictionaries uses curly braces and colons to signify keys and their associated values.

eg:-  {'key1':'value1','key2':'value2','key3':'value3'}

2)Tuples(tuple) ---> tuples are orderd immutable sequence of objects, once an element is inside a tuple, it can not be reassigned. Tuples use parenthesis. 

eg:- (1,2,3)

3)Sets(set) ---> Sets are unorderd collections of unique elements it means that there can be only one representative of the same object.

eg:- {'a','b'}
