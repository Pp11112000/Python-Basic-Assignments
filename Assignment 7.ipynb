## 10.In regular expressions, what is the difference between the + and * characters?

a) + character ---> This character specifies an expression to its left for one or more times.

Example:- s+ is an expression which gives "s", "ss", "sss" and so on.

b) * character ---> This character specifies an expression to its left for 0 or more times.

## 14. What is the difference between .*  and .* ?

a) .* ---> The .* uses greedy mode. It will always try to match as much text as possible.

b) .* ? ---> The .* ? uses non-greedy mode. It will match any and all text in non-greedy fashion. Like with braces, the question mark tells python to match in non-greedy way.

```python
# Example of .*
GreedyRegex = re.compile(r'<.*>')
search = GreedyRegex.search('< Hi Pranit Arun Patil> how are you>')
search.group()
```

```python
# Example of .*?
GreedyRegex = re.compile(r'<.*?>')
search = GreedyRegex.search('<Hi Pranit Arun Patil> how are you>')
search.group()
```

## 20. How would you write a regex that match a number with comma for every three digits? It must match the given following:
'42'

'1,234'

'6,368,745'

but not the following:

'12,34,567' (which has only two digits between the commas)

'1234' (which lacks commas)


```python
Regex1 = re.compile(r'^\d{1,3}(,\d{3})*$')
s1 = Regex1.search('42')
s1.group()
```

```python
Regex2 = re.compile(r'^\d{1,3}(,\d{3})*$')
s2 = Regex2.search('1,234')
s2.group()
```

```python
Regex3 = re.compile(r'^\d{1,3}(,\d{3})*$')
s3 = Regex3.search('6,368,745')
s3.group()
```
