## 7. What is a shortcut for the following code?
if 'color' not in spam:

spam['color'] = 'black'



The shortcut of above code is written below by using setdefault() method. By using this method in which we have to give key-value pair so that when we create a dictionary by default it will be present in that created dictionary. 

```python
spam = {'cat' : 97}
spam.setdefault('color', 'black')
spam
```
