```python
10. How can you tell the difference between break and continue?

Ans --->
```

a) break --> The break statement is used to terminate the loop or statement in which it is present. After that, the control will pass to the statements that are present after the break statement, if available. If the break statement is present in the nested loop, then it terminates only those loops which contains break statement.

b)continue --> Continue is also a loop control statement just like the break statement. continue statement is opposite to that of break statement, instead of terminating the loop, it forces to execute the next iteration of the loop.
As the name suggests the continue statement forces the loop to continue or execute the next iteration. When the continue statement is executed in the loop, the code inside the loop following the continue statement will be skipped and the next iteration of the loop will begin.

```python
#Example of break statement
for num in range(0,11,1):
    if num == 5:
        break
    print(num)
```

```python
#Example of continue statement
for num in range(0,11,1):
    if num == 6:
        continue
    print(num)
```

```python
11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Ans --->
```

a) range(10) ---> range(stop) ---> When user call range() with one argument, user will get a series of numbers that starts at 0 and includes every whole number up to, but not including, the number that user have provided as the stop. 

For Example:- 



```python
for i in range(10):
    print(i, end = " ")
```

b) range(0, 10) ---> range(start, stop) ---> When user call range() with two arguments, user get to decide not only where the series of numbers stops but also where it starts, so user don’t have to start at 0 all the time. User can use range() to generate a series of numbers from A to B using a range(A, B). 

For Example - 

```python
for i in range(5,10):
    print(i, end = " ")
```

c) range(0, 10, 1) ---> range(start, stop, step) ---> When the user call range() with three arguments, the user can choose not only where the series of numbers will start and stop but also how big the difference will be between one number and the next. If the user doesn’t provide a step, then range() will automatically behave as if the step is 1. 

For Example -

```python
for i in range(5,10,2):
    print(i, end = " ")
```
