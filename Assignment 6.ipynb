## 1. What are escape characters, and how do you use them?

Escape character consists of backslash ( \ ) followed by the character you want to add to the string.

There are several different different escape characters.

a) \ ' ---> Single Quote

b) \ '' ---> Double Quote

c) \t ---> tab

d) \n ---> new line

## 2. What do the escape characters n and t stand for?

a) \n ---> new line

b) \t ---> tab

```python
# example of new line (\n)
print('hi hello \nhow are you')
```

```python
# example of tab (\t)
print('hi hello \thow are you')
```

## 9. What are the methods for right-justifying, left-justifying, and centering a string?

a) right-justifying ---> rjust()

b) left-justifying ---> ljust()

c) centering ---> centre()

```python
mystr = 'how are you'
```

```python
# right-justifying
mystr.rjust(100)
```

```python
# left-justifying
mystr.ljust(100)
```

```python
# centering
mystr.center(100)
```
